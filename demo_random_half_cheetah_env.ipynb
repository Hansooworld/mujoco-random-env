#### Import Libraries

```python
import glfw, os, mujoco_py, warnings, math
warnings.filterwarnings("ignore") # ignore warnings
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from IPython.display import display, HTML
from class_half_cheetah import HalfCheetahRandomEnvClass, HalfCheetahRandomEnvClassWithBox, HalfCheetahRandomEnvClassMixVersion
%matplotlib inline
print ("Done.")
```

#### Random Leg's weight

```python
max_tick, plot_every = 500, 100
for tick in range(max_tick):
    if (tick+1) % plot_every == 0:
        env = HalfCheetahRandomEnvClass(xml_path='xml/half_cheetah_leg.xml', rand_mass=[1,5], rand_fric=None, render_mode="rgb_array",VERBOSE=False)
        for i in range(10):
            env.step(np.random.standard_normal(6)*0.1)        
        frame = env.render_center()
        plt.figure(figsize=(env.render_w/env.render_res, env.render_h/env.render_res))
        plt.imshow(frame)
        plt.title("Leg Weight [{}]".format(env.get_leg_weight()), fontsize=18)
        plt.show()
```

```python
max_tick, plot_every = 500, 100
for tick in range(max_tick):
    if (tick+1) % plot_every == 0:
        env = HalfCheetahRandomEnvClass(xml_path='xml/half_cheetah_leg.xml', rand_mass=[1,5], rand_fric=None, render_mode="rgb_array",VERBOSE=False)
        env.set_leg_weight(tick/100)
        for i in range(10):
            env.step(np.random.standard_normal(6)*0.1)        
        frame = env.render('rgb_array')
        plt.figure(figsize=(env.render_w/env.render_res, env.render_h/env.render_res))
        plt.imshow(frame)
        plt.title("Leg Weight [{}]".format(env.get_leg_weight()), fontsize=18)
        plt.show()
```

#### Random Box's weight, Leg's weight, Friction

```python
max_tick, plot_every = 500, 100
env = HalfCheetahRandomEnvClassMixVersion(xml_path='xml/half_cheetah_box_leg.xml', rand_mass_box=[1,5], rand_mass_leg=[1,5], rand_fric=[0.2, 2], render_mode="rgb_array",VERBOSE=False)
for tick in range(max_tick):
    if (tick+1) % plot_every == 0:
        env.reset()
        env.reset_random()
        for i in range(10):
            env.step(np.random.standard_normal(6)*0.1)
        frame = env.render_center()
        plt.figure(figsize=(env.render_w/env.render_res, env.render_h/env.render_res))
        plt.imshow(frame)
        plt.title("Box Weight [{}] Leg Weight [{}] Friction [{}]".format(env.get_box_weight(), env.get_leg_weight(), env.get_fric()), fontsize=12)
        plt.show()
```

### Set Box's weight, Leg's weight, Friction

```python
max_tick, plot_every = 500, 100
env = HalfCheetahRandomEnvClassMixVersion(xml_path='xml/half_cheetah_box_leg.xml', rand_mass_box=[1,5], rand_mass_leg=[1,5], rand_fric=[0.2, 2], render_mode="rgb_array",VERBOSE=False)
for tick in range(max_tick):
    if (tick+1) % plot_every == 0:
        env.set_box_weight(tick/100)
        env.set_leg_weight(tick/100)
        env.set_fric(tick/100)
        for i in range(10):
            env.step(np.random.standard_normal(6)*0.1)
        frame = env.render(mode='rgb_array', width=2000, height=1000)
        plt.figure(figsize=(env.render_w/env.render_res, env.render_h/env.render_res))
        plt.imshow(frame)
        plt.title("Box Weight [{}] Leg Weight [{}] Friction [{}]".format(env.get_box_weight(), env.get_leg_weight(), env.get_fric()), fontsize=12)
        plt.show()
```
