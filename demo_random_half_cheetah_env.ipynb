#### Import Libraries

```python
import glfw, os, mujoco_py, warnings, math
warnings.filterwarnings("ignore") # ignore warnings
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from IPython.display import display, HTML
from class_half_cheetah import HalfCheetahRandomEnvClass, HalfCheetahRandomEnvClassWithBox
%matplotlib inline
print ("Done.")
```

#### Random Leg's weight

```python
max_tick, plot_every = 500, 100
for tick in range(max_tick):
    if (tick+1) % plot_every == 0:
        env = HalfCheetahRandomEnvClass(rand_mass=[1,5], rand_fric=None, render_mode="rgb_array",VERBOSE=False)
        for i in range(10):
            env.step(np.random.standard_normal(6)*0.1)        
        frame = env.render_center()
        plt.figure(figsize=(env.render_w/env.render_res, env.render_h/env.render_res))
        plt.imshow(frame)
        plt.title("Leg Weight [{}]".format(env.get_leg_weight()), fontsize=18)
        plt.show()
```

### Random Friction

```python
for i in range(5):
    env = HalfCheetahRandomEnvClass(rand_mass=None, rand_fric=[0.2, 1.0], render_mode="rgb_array",VERBOSE=False)
    print("Env Friction: [{}]".format(env.get_fric()))
```
